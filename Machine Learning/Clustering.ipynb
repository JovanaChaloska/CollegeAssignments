```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('Data_Cortex_Nuclear.xls')
df = df.iloc[:, 1:-4]
pd.set_option('display.max_rows', None)
null_counts = df.isnull().sum()
print(null_counts)
pd.reset_option('display.max_rows') #za vidam sekoj atribut kolku null vrednosti ima, koga gi dropnuvam tie koloni mi se prepolovuva podatocnoto mnozestvo
print(null_counts)
df 
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = df.dropna()
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_scaled
```

```python
pca = PCA(n_components=2)  
df_pca = pca.fit_transform(df_scaled)
df_pca = pd.DataFrame(data=df_pca, columns=['PC1', 'PC2'])
```

```python
plt.scatter(df_pca['PC1'], df_pca['PC2'])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
pca = PCA(n_components=3) 
df_pca3 = pca.fit_transform(df_scaled)
df_pca3 = pd.DataFrame(data=df_pca3, columns=['PC1', 'PC2', 'PC3'])

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_pca3['PC1'], df_pca3['PC2'], df_pca3['PC3'])
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.manifold import TSNE

for cov_type in ['spherical', 'diag', 'tied', 'full']:
    gmm = GaussianMixture(n_components=3, covariance_type=cov_type) 
    gmm.fit(df_scaled)
    labels = gmm.predict(df_scaled)

    tsne = TSNE(n_components=3, random_state=42)
    df_tsne = tsne.fit_transform(df_scaled)

    df_tsne = pd.DataFrame(data=df_tsne, columns=['D1', 'D2', 'D3'])
    df_tsne['Cluster'] = labels

    plt.figure(figsize=(10, 8))
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(df_tsne['D1'], df_tsne['D2'], df_tsne['D3'], c=df_tsne['Cluster'])
    ax.set_title('t-SNE Visualization of GMM Clusters with ' + cov_type + ' covariance matrix')
    ax.text(0.05, 0.9, 0, 'BIC score: %.1f' % gmm.bic(df_scaled), transform=ax.transAxes)
    plt.show()
```

```python
from scipy import stats

z_scores = stats.zscore(df)
df_no_outliers = df[(z_scores < 3).all(axis=1)]
scaler = StandardScaler()
df_scaled_no_outliers = scaler.fit_transform(df_no_outliers)

for cov_type in ['spherical', 'diag', 'tied', 'full']:
    gmm = GaussianMixture(n_components=3, covariance_type=cov_type) 
    gmm.fit(df_scaled_no_outliers)
    labels = gmm.predict(df_scaled_no_outliers)

    tsne = TSNE(n_components=3, random_state=42)
    df_tsne = tsne.fit_transform(df_scaled_no_outliers)

    df_tsne = pd.DataFrame(data=df_tsne, columns=['D1', 'D2', 'D3'])
    df_tsne['Cluster'] = labels

    plt.figure(figsize=(10, 8))
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(df_tsne['D1'], df_tsne['D2'], df_tsne['D3'], c=df_tsne['Cluster'])
    ax.set_title('t-SNE Visualization of GMM Clusters with ' + cov_type + ' covariance matrix')
    ax.text(0.05, 0.9, 0, 'BIC score: %.1f' % gmm.bic(df_scaled_no_outliers), transform=ax.transAxes)
    plt.show()
```

```python
for cov_type in ['spherical', 'diag', 'tied', 'full']:
    gmm = GaussianMixture(n_components=2, covariance_type=cov_type) 
    gmm.fit(df_scaled_no_outliers)
    labels = gmm.predict(df_scaled_no_outliers)

    tsne = TSNE(n_components=3, random_state=42)
    df_tsne = tsne.fit_transform(df_scaled_no_outliers)

    df_tsne = pd.DataFrame(data=df_tsne, columns=['D1', 'D2', 'D3'])
    df_tsne['Cluster'] = labels

    plt.figure(figsize=(10, 8))
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(df_tsne['D1'], df_tsne['D2'], df_tsne['D3'], c=df_tsne['Cluster'])
    ax.set_title('t-SNE Visualization of GMM Clusters with ' + cov_type + ' covariance matrix')
    ax.text(0.05, 0.9, 0, 'BIC score: %.1f' % gmm.bic(df_scaled_no_outliers), transform=ax.transAxes)
    plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

cluster = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
labels = cluster.fit_predict(df_scaled)

silhouette_avg = silhouette_score(df_scaled, labels)
silhouette_avg
```
